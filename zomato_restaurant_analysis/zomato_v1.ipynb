### Rate

```python
zomato.rate.unique()
```

**NOTE**:-
- So, Now we found this column has **'NEW' ,"-"** values which should replaced by np.nan
- there is /5 as extra and no need for it
- values represents as strings instead of float


```python
zomato.rate.replace(('NEW','-'),np.nan,inplace =True)
# first make it as string
zomato.rate = zomato.rate.astype('str')
# remove the "/5" 
zomato.rate = zomato.rate.apply(lambda x: x.replace('/5','').strip())
# convert column type to float
zomato.rate = zomato.rate.astype('float')
```
